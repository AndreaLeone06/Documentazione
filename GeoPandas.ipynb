```python
%matplotlib inline
import pandas as pd
import geopandas
```

```python
countries = geopandas.read_file("https://github.com/jorisvandenbossche/geopandas-tutorial/blob/main/data/ne_110m_admin_0_countries.zip?raw=true")#importazione file
countries
```

```python
countries.plot() #visualizzazione grafico
```

```python
africa = countries[countries['continent'] == 'Africa']
africa.plot()
```

```python
countries[countries['name'] == 'Italy'].area #mostra l'area
```

```python
cities = geopandas.read_file("https://github.com/jorisvandenbossche/geopandas-tutorial/blob/main/data/ne_110m_populated_places.zip?raw=true")
cities
```

```python
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15, 10)) #edge color= colore bordo,    face color= colore sfondo
cities.plot(ax=ax, color='red') #posiziona cities 
ax.set(xlim=(-20, 60), ylim=(-40, 40)) #limitazione di dove visualizzare il tutto
```

```python
countries[countries['name'] == 'Italy'].plot()
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15, 10))
cities.plot(ax=ax, color='red')
ax.set(xlim=(6.7, 18.5), ylim=(36.6, 47.15))
```

```python
import contextily #visualizzare le mappe che andremo a mettere sotto le informazioni
ax = stations.plot(figsize=(12,6), markersize=5) #markersize= dimensione pallini
contextily.add_basemap(ax) #*
```

```python
ax = stations.plot(figsize=(24, 12), column='available_bikes', legend=True, markersize= 20) #legenda= "columns="
contextily.add_basemap(ax)
```

```python
districts = geopandas.read_file("https://raw.githubusercontent.com/jorisvandenbossche/geopandas-tutorial/main/data/paris_districts_utm.geojson")
districts
```

```python
ax= districts.plot(edgecolor='k', facecolor='none', figsize=(15, 10))
contextily.add_basemap(ax)
```

```python
districts['area'] = districts.geometry.area / 1e6 #crea una colonna, con la divisione alla fine per scalare
districts
```

```python
districts['population_density'] = districts['population'] / districts.geometry.area * 10**6
districts.plot(column='population_density', figsize=(12, 6), legend=True)
```
