caricare in una lista i primi 100 numeri interi, visualizzare la lista e caricare in un'altra lista solo i multipli di 5 presenti nella prima lista.

```python
lista1 = [numero for numero in range(101)]
print(lista1)
lista2 = []
for elemento in lista1:
  if elemento % 5 == 0:
    lista2.append(elemento)
print(lista2)
```

generare 100 numeri casuali interi da memorizzare in una lista. creare poi altre 2 liste, una contenente i valori pari, e l'altra i valori dispari.

```python
list1 = [random.randint(1,100) for i in range(100)]
list2 = []
list3 = []
for elemento in list1:
  if elemento % 2 == 0:
    list2.append(elemento)
  else: list3.append(elemento)
  print(list1)
  print(list2)
  print(list3)
```
